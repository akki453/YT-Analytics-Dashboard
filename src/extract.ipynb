```python
#Get top 50 videos trending in a region

from googleapiclient.discovery import build

api_key = ''

youtube = build('youtube', 'v3', developerKey=api_key)

def get_top_videos(region='IN', max_results=50):
    request = youtube.videos().list(
        part='snippet,statistics',
        chart='mostPopular',
        regionCode=region,
        maxResults=min(max_results, 50)
    )
    response = request.execute()
    
    top_videos = []
    for item in response['items']:
        title = item['snippet']['title']
        video_id = item['id']
        views = int(item['statistics'].get('viewCount', 0))
        top_videos.append((title, video_id, views))
    
    return sorted(top_videos, key=lambda x: -x[2]) 

videos = get_top_videos('US')  # or 'IN' for India
for title, vid, views in videos:
    print(f"{title} ({views:,} views) - https://youtube.com/watch?v={vid}")
```

```python
#list of all videos from a channel by keyword
from googleapiclient.discovery import build

api_key = ''
youtube = build('youtube', 'v3', developerKey=api_key)

def get_first_channel_id(keyword):
    """Find the first YouTube channel that matches a keyword"""
    request = youtube.search().list(
        part='snippet',
        q=keyword,
        type='channel',
        maxResults=1
    )
    response = request.execute()

    if response['items']:
        return response['items'][0]['snippet']['channelId']
    else:
        return None

def get_uploads_playlist_id(channel_id):
    """Get the 'uploads' playlist ID for a channel"""
    request = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    )
    response = request.execute()
    
    return response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

def get_all_videos_from_playlist(playlist_id):
    """Fetch all videos from a playlist (the uploads playlist), including publish date"""
    videos = []
    next_page_token = None

    while True:
        request = youtube.playlistItems().list(
            part='snippet',
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response['items']:
            snippet = item['snippet']
            video_id = snippet['resourceId']['videoId']
            title = snippet['title']
            published_at = snippet['publishedAt']
            videos.append({
                'title': title,
                'video_id': video_id,
                'published_at': published_at
            })

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return videos

# ==== Example usage ====
keyword = "penguinz0"
channel_id = get_first_channel_id(keyword)

if channel_id:
    print(f"âœ… First matching channel ID: {channel_id}")
    playlist_id = get_uploads_playlist_id(channel_id)
    videos = get_all_videos_from_playlist(playlist_id)

    print(f"\nðŸŽ¥ Found {len(videos)} videos:")
    for video in videos:
        print(f"{video['published_at']} - {video['title']} - https://youtube.com/watch?v={video['video_id']}")
else:
    print("âŒ No channel found for that keyword.")

```

```python
#Channel metadata by username
from googleapiclient.discovery import build

api_key = ''
youtube = build('youtube', 'v3', developerKey=api_key)

def get_channel_metadata_by_username(username):
    """Fetch channel metadata using YouTube legacy username"""
    request = youtube.channels().list(
        part='snippet,statistics',
        forUsername=username
    )
    response = request.execute()

    if not response['items']:
        print("âŒ Channel not found with username:", username)
        return None

    data = response['items'][0]
    snippet = data['snippet']
    stats = data['statistics']

    metadata = {
        'channel_id': data['id'],
        'title': snippet.get('title'),
        'description': snippet.get('description'),
        'published_at': snippet.get('publishedAt'),
        'country': snippet.get('country', 'N/A'),
        'subscriber_count': stats.get('subscriberCount'),
        'view_count': stats.get('viewCount'),
        'video_count': stats.get('videoCount')
    }

    return metadata

# ==== Example Usage ====

username = 'penguinz0'  # Example of a legacy YouTube username
info = get_channel_metadata_by_username(username)

if info:
    print("\nðŸ“º Channel Metadata:")
    for key, value in info.items():
        print(f"{key}: {value}")
```

```python
#Get channel stats by keyword


from googleapiclient.discovery import build
api_key = ''
youtube = build('youtube', 'v3', developerKey=api_key)

def get_first_channel_by_keyword(keyword):
    """Search YouTube for a keyword and return the first matching channel's ID"""
    request = youtube.search().list(
        part='snippet',
        q=keyword,
        type='channel',
        maxResults=1
    )
    response = request.execute()

    if not response['items']:
        print(f"âŒ No channel found for keyword: {keyword}")
        return None

    item = response['items'][0]
    channel_id = item['snippet']['channelId']
    return channel_id

def get_channel_stats_by_id(channel_id):
    """Fetch statistics and info for a channel using its channel ID"""
    request = youtube.channels().list(
        part='snippet,statistics',
        id=channel_id
    )
    response = request.execute()

    if not response['items']:
        print(f"âŒ Channel not found for ID: {channel_id}")
        return None

    data = response['items'][0]
    snippet = data['snippet']
    stats = data['statistics']

    metadata = {
        'Channel Title': snippet.get('title'),
        'Channel ID': data.get('id'),
        'Description': snippet.get('description'),
        'Subscribers': stats.get('subscriberCount'),
        'Total Views': stats.get('viewCount'),
        'Total Videos': stats.get('videoCount'),
        'Created On': snippet.get('publishedAt'),
        'Country': snippet.get('country', 'N/A')
    }

    return metadata

# ==== Example Usage ====
keyword = "python programming"
channel_id = get_first_channel_by_keyword(keyword)

if channel_id:
    stats = get_channel_stats_by_id(channel_id)
    if stats:
        print("\nðŸ“Š Channel Statistics:")
        for key, value in stats.items():
            print(f"{key}: {value}")
else:
    print("âŒ Could not find a channel from the keyword.")
```
